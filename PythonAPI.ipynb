```python
import requests

response = requests.get("https://api.quotable.io/random", verify=False)

if response.status_code == 200:
    data = response.json()
    quote = data["content"]
    author = data["author"]

    print("\n✨ Quote of the Day ✨")
    print(f"\"{quote}\"")
    print(f"— {author}\n")
else:
    print("Failed to fetch quote!")
```
